## Conversion of Electricity Power usage to Carbon emissions

From https://www.epa.gov/energy/greenhouse-gases-equivalencies-calculator-calculations-and-references, we have the conversion of Energy used in KWh to Carbon emissions to be: 

$$\text{CO₂ emissions (metric tons)} = \text{Power consumption (kWh)} \times \text{CO₂ intensity (kg CO₂ per kWh)} \div 1000$$

So say, In 2017, California used about 3,200 Kwh of power, We have the equivalent carbon emissions to be: 
We're gonna use US average of 0.42kg CO2/KWh



```python
Avg_E_Ca_Res = 3200
Carbon =3200 * 0.42* Avg_E_Ca_Res / 100
print(Carbon)
```

## Equivalence of Miles travelled to emissions


$$\text{CO₂ emissions (g/mile)} = \text{Fuel emission factor} \times \text{Fuel consumption per mile}$$
Common emission factors in grams of CO₂ per mile (gCO₂/mi) are:

1. Passenger car: ~404 gCO₂/mi 
2. Light-duty trucks (e.g., SUVs): ~500 gCO₂/mi 
3. Heavy-duty trucks: ~1000 gCO₂/mi
4. Airplane (short-haul flights): ~250 gCO₂/mi (per passenger)


$$\text{Total CO₂ (g)} = \text{Miles traveled} \times \text{CO₂ emission factor (g/mile)}$$

$$\text{Total CO₂ (metric tons)} = \frac{\text{Total CO₂ (g)}}{1,000,000}$$

So we get user input in terms of distance travelled, and we convert it Total Co2



```python
#We will assume that the mode of transportation is the average of Car and Airplane 

average_emission_factor = 327  # gCO₂/mi
Miles_travelled = 12500 # average in california in a year
Co2 = Miles_travelled * average_emission_factor / 1000000
print(Co2)
```
